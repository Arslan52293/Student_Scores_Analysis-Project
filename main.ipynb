```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#Drop Unammes col

```python
df = df.drop("Unnamed: 0" , axis = 1)
print(df.head())
```

#Changed the coloumn data

```python
df["EthnicGroup"] = df["EthnicGroup"].str.replace("group C" , "Group C")
df.head()
```

#gender distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#the number of females in the data is more than males 

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb , annot = True)
plt.title("Relationship b/w Parent educations and students scores")
plt.show()
```

# from the above chart we have concluded that the education of parent's have a good impact on there children scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1 , annot = True)
plt.title("Relationship b/w Parent-Martial-Status and Student-Scores ")
plt.show()
```

```python
gb2 = df.groupby("WklyStudyHours").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
print(gb2)
```

```python
sns.heatmap(gb2 , annot = True)
plt.title("Relationship b/w Weekly-Study-Hours and Student-Scores ")
plt.show()
```

```python
#from the above chart Weekly study hours would impact on students scores
```

Distribution of ethnic group

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "Group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

label = ["group A" , "group B" , "group C" , "group D" , "group E"]
mlst = [groupA["EthnicGroup"] , groupB["EthnicGroup"] , groupC["EthnicGroup"] , groupD["EthnicGroup"] , groupE["EthnicGroup"]]
plt.pie(mlst , labels = label , autopct = "%1.2f%%")
plt.title("Distribution of Ethic Groups")
plt.show()
```
